```python
import random

# User is invited to decide the number of time to repeat experiment
# Roll two six-sided dice.
# Iterate the roll six times, counting the sums.
answer = int(input("How many rolls do you want?:    "))
roll_count = 0
grand_total = 0
roll_average = 0

while roll_count < int(answer):
    first_die = random.randint(1, 6)
    second_die = random.randint(1, 6)
    dice_sum = first_die + second_die
    roll_count = roll_count + 1
    grand_total = grand_total + dice_sum
    roll_average = grand_total / roll_count
    roll_average = round(roll_average, 0)

    print("You rolled a " + str(dice_sum) + "!")

    print("This is the sum of your roll number " + str(roll_count) + "!")

#  Here is the grand total of all the rolls.
print("This is your grand total " + str(grand_total) + "!")
print("This is your roll average " + str(roll_average) + "!")
```
